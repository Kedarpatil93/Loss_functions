```python
import numpy as np
import matplotlib.pyplot as plt

# Define x range
x = np.linspace(-5, 5, 400)
y_true = 1  # True value for loss functions

# Loss Functions
mse = (y_true - x) ** 2
mse_deriv = 2 * (x - y_true)
mae = np.abs(y_true - x)
mae_deriv = np.where(x < y_true, -1, np.where(x > y_true, 1, 0))

# Activation Functions
sigmoid = 1 / (1 + np.exp(-x))
sigmoid_deriv = sigmoid * (1 - sigmoid)
tanh = np.tanh(x)
tanh_deriv = 1 - tanh ** 2
relu = np.maximum(0, x)
relu_deriv = np.where(x > 0, 1, 0)
# Softmax simplified for single input (approximation for visualization)
softmax_base = np.exp(x) / (np.exp(x).sum())  # Approx. for single input
softmax_deriv = softmax_base * (1 - softmax_base)  # Approx. derivative

# Create a 3x2 grid of subplots
plt.figure(figsize=(15, 10))

# MSE
plt.subplot(3, 2, 1)
plt.plot(x, mse, label='MSE', color='blue')
plt.plot(x, mse_deriv, label='MSE Derivative', linestyle='--', color='orange')
plt.title('Mean Squared Error')
plt.xlabel('Prediction Error')
plt.ylabel('Value')
plt.legend()
plt.grid(True)

# MAE
plt.subplot(3, 2, 2)
plt.plot(x, mae, label='MAE', color='green')
plt.plot(x, mae_deriv, label='MAE Derivative', linestyle='--', color='red')
plt.title('Mean Absolute Error')
plt.xlabel('Prediction Error')
plt.ylabel('Value')
plt.legend()
plt.grid(True)

# Sigmoid
plt.subplot(3, 2, 3)
plt.plot(x, sigmoid, label='Sigmoid', color='purple')
plt.plot(x, sigmoid_deriv, label='Sigmoid Derivative', linestyle='--', color='brown')
plt.title('Sigmoid')
plt.xlabel('Input')
plt.ylabel('Output / Derivative')
plt.legend()
plt.grid(True)

# Softmax (Approximated)
plt.subplot(3, 2, 4)
plt.plot(x, softmax_base, label='Softmax (Approx)', color='cyan')
plt.plot(x, softmax_deriv, label='Softmax Derivative (Approx)', linestyle='--', color='magenta')
plt.title('Softmax (Single Input Approx)')
plt.xlabel('Input')
plt.ylabel('Output / Derivative')
plt.legend()
plt.grid(True)

# Tanh
plt.subplot(3, 2, 5)
plt.plot(x, tanh, label='Tanh', color='teal')
plt.plot(x, tanh_deriv, label='Tanh Derivative', linestyle='--', color='pink')
plt.title('Tanh')
plt.xlabel('Input')
plt.ylabel('Output / Derivative')
plt.legend()
plt.grid(True)

# ReLU
plt.subplot(3, 2, 6)
plt.plot(x, relu, label='ReLU', color='gray')
plt.plot(x, relu_deriv, label='ReLU Derivative', linestyle='--', color='yellow')
plt.title('ReLU')
plt.xlabel('Input')
plt.ylabel('Output / Derivative')
plt.legend()
plt.grid(True)

# Adjust layout and save
plt.tight_layout()
plt.savefig('functions_and_derivatives.png')
plt.show()

print("Plots generated and saved as 'functions_and_derivatives.png' in the current directory.")
```
